```python
import pandas as pd
import numpy as np
import requests as req
import time
import datetime
import json
```

```python
url = "https://data.austintexas.gov/resource/r3af-2r8x.json?$limit=50000&$offset=0"
```

```python
traffic_response = req.get(url)
traffic_json = traffic_response.json()
traffic_json
```

```python
df = pd.DataFrame(traffic_json)
len(df)
```

```python
clean_df = df[['address','issue_reported','location_latitude', 'location_longitude', 'published_date']]
clean_df.head()
```

```python
print(df['published_date'][0], df['traffic_report_status_date_time'][0])
```

```python
dropped = clean_df.dropna()
newdate = dropped.loc[0:, 'published_date']
listy = []
for i in newdate:
    listy.append(i[0:10])
# newdate

print(listy)
```

```python

dropped['published_date'].replace(listy)
dropped.head()
# for x in listy:
#     dropped['published_date'] = x
# dropped.head()
```
