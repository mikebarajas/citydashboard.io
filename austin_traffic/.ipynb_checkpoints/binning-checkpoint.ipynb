```python
def json_scrape():
    url = "https://data.austintexas.gov/resource/r3af-2r8x.json?$limit=50000&$offset=0"

    # get the json from austin data api
    traffic_response = req.get(url)
    traffic_json = traffic_response.json()
    time.sleep(1)

    # append json into DataFrame
    df = pd.DataFrame(traffic_json)
    time.sleep(1)

    # Selecting specific columns needed
    clean_df = df[['address','issue_reported','location_latitude', 'location_longitude', 'published_date']]

    # Drop any NAN values and append to a list
    dropped = clean_df.dropna()
    newdate = dropped.loc[0:, 'published_date']
    listy = []
    for i in newdate:
        listy.append(i[0:10])
    time.sleep(1)

    #  delete old values for published_date and pass in new list to published_date column
    xyz = dropped.drop('published_date', axis=1)
    xyz['published_date'] = listy
    time.sleep(1)

    return xyz
```

```python
from flask import Flask, render_template, jsonify
from flask_pymongo import PyMongo
import os
import time
import datetime
import json, sys
import pandas as pd
import requests as req
from bson import json_util
import numpy as np
import re
```

```python
df = pd.DataFrame(json_scrape())
df.head()
```

```python
df['published_date'] = df['published_date'].str.replace('-',',')
```

```python
df['published_date'] = df['published_date'].str.replace(',','')
```

```python
df.head()
```

```python
df_date = df.sort_values('published_date')
df_date.head()

```

```python
df_date[["published_date"]] = df_date[["published_date"]].astype(int)
df_date.head()
```

```python
df_bins = [20170925, 20171031, 20171130, 20171231, 20180131, 20180228, 20180331]
df_labels = ["October_2017", "November_2017", "December_2017", "January_2018", "February_2018", "March_2018"]
df_date["month"] = pd.cut(df_date["published_date"], df_bins, labels=df_labels)
```

```python
df_date.head()
```

```python
df_date[["published_date"]] = df_date[["published_date"]].astype(str)
```

```python
df_date.head()
```

```python
group_df = df_date.groupby(["month"])
group_ct = group_df["issue_reported"].count()
```

```python
groups_df = df_date.groupby(["month","issue_reported"])
groups_ct = groups_df["issue_reported"].count()
```
